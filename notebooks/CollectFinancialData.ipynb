In this notebook, I demonstrate how to collect financial data from the investing.com website through the investpy library.

Financial instruments analyzed:

1.   Stocks;

3.   ETFs;

6.   Bonds;
7.   Commodities;
7.   Cryptocurrencies.

Some investpy functions require the country in which the financial asset is traded. To collect historical data about a particular asset, it will be necessary to inform the period (in days).

```python
country="brazil"
```

```python
initial_date = "10/05/2019"
```

```python
final_date = "08/05/2020"
```

# Stocks



Source: https://www.investing.com/equities/

Gets countries available for stock query.

```python
stock_countries = investpy.get_stock_countries()
```

```python
print(stock_countries)
```

Lists the shares traded in a specific country.

```python
stocks = investpy.get_stocks_list(country=country)
```

```python
print("Stocks in brazil market:\n{stocks}".format(stocks=stocks))
```

Collects an overview of the stocks traded in a given country.

```python
stocks_overview = investpy.get_stocks_overview(country="brazil")
```

```python
print(stocks_overview)
```

Collects information about a given stock.

```python
stock_information = investpy.get_stock_information(stock="PETR3", country="brazil")
```

```python
print(stock_information)
```
